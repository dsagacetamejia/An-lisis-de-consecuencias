## Desarrollo y ecuaciones

Los datos experimentales indican que los efectos combinados del combustible reactivo, la densidad de obstáculos  y el confinamiento puede correlacionarse  con la velocidad de la flama.

La Tabla 1  describe 27 posibles combinaciones de la máxima velocidad de la flama, basados en las dimensiones  de expanción de la flama y ésta velocidad esta expresada en unidades del Número de Mach. La reactividad esta clasificada como baja, media y alta según las recomendaciones del TNO.

El Metano y el Monoxido de carbono son considerdos de baja reactividad.El Hidrógeno, Acetileno, Etileno , Oxido de etileno y Oxido de propileno son considerados de alta react.Los otros combustibles estan consideramos de media reactividad.
Para mezclas de combustibles se clasifica de acuerdo a la concentración del componente más reactivo.



El confinamiento esta basado en tres simentrías.
- Punto-Simétrico: 

La flama se expande esféricamente desde su punto de ignición.La superficie de la flama se incrementa  con el cuadrado de la distancia desde el punto de ignición.



- Linea-Simétrica: 


Es un flama cilindrica entre dos planos.
El área superficial de la flama es proporcional a la distancia del punto de ignición.

- Plano-Simétrico: 



Es una flama plana en un tubo. El área superficial proyectada de la flama es constante 


La densidad de obstaculos es función de la relación de bloqueo y pitch. La relación de bloqueo esta definida por el área  transversal bloqueada por los obstaculos. Se clasifican en:


- Baja: 

La relación de bloque menor que el 10%, asume que hay pocos obstaculos en el camino de la flama y sólo hay una o dos láminas de obstaculos.

- Media : 

Cae entre estas dos categorias, es decir, la relación de bloqueo esta entre (10-40)% y puede haber dos o tres láminas.

- Alta : 

Cuando hay tres o más láminas muy cercanas con una relación de bloqueo mayor de 40% por lámina

Después de definir la flama, se calcula una distancia escalada (D)

- Ecuación 1.  $D=\frac{R}{(\frac{E}{P_0})^{1/3}}$

Posteriormente se obtienen los valores de sobrepresión e impulso utilizando las gráficas del modelo Baker-Strehlow.

## Resultados


```python
#Datos

R=30 #m
v= 344 #m/s
P0= 101325 #Pa
E= 14600 #MJ
```

Primero sedetermina la velocidad de la flama utilizando  la TABLA 1.

La nube de vapor esta confinada en un tanque resistente y la flama sólo puede expanderse en dos direcciones, se considera una densidad  de obstaculos media y como el combustible es propano la reactividad es media.

```python
Mw=0.662
print("No. Match= "+ str(round(Mw,2))+" ")
```

Obteniendo distancia escalada para utilizar el gráfico correspondiente:

Cálculo de la Distancia  Sachs-escalada, para después obtener los valores de sobrepresión e impulso utilizando la gráfica 1 para la sobrepresión y la gráfica 2 pára el impulso


```python
D=R/(E*1000000/P0)**(1/3)

print("D= "+ str(round(D,2))+" ")
```

Se calcula la sobrepresión escalada a través de una interpolación:

```python
import numpy as np
Ps = 0.2017*np.log(Mw) + 0.695
Ps_now=Ps*P0*0.001
Imp=0.0002*Mw**5 - 0.004*Mw**4 + 0.0322*Mw**3 - 0.1212*Mw**2 + 0.2086*Mw - 0.0286
Imp_now=Imp*P0**(2/3)*(E*10**6)**(1/3)/(v)
print("Sobre presión escalar = "+ str(round(Ps,3))+" ")
print("Sobre presión actual = "+ str(round(Ps_now,3))+" kPa ")
print("Impulso escalado = "+ str(round(Imp,3))+" ")
print("Impulso  = "+ str(round(Imp_now,3))+" KPa-ms")

```
