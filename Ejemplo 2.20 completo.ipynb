### Consideraciones 
- Se asume que existe una equivalencia entre un material inflamable y el TNT
- La eficiencia empirica se considera 0.5 como promedio
- El calor de combustión del TNT es de 4 437 - 4 765 kJ/kg, pero se considera un promedio de 4 652 kJ/kg










## Desarrollo y ecuaciones

Se asume que existe una equivalencia entre un material inflamable y TNT a través de la siguiente relación:

- Ecuación 1.  $W=\frac{\eta ME_c}{E_{TNT}} $


Se usa un parámetro para escalar la sobrepresión:


- Ecuación 2. $P_s=\frac{P^0}{P_a}$

Con el valor de Ps se obtiene un parámetro Z que involucra una relación distancia/masa.

Este valor se obtiene a través de la siguiente gráfica:

Con los datos anteriores, ya se puede calcular la distancia:

- Ecuación 3. $R=ZW^{1/3}$



## Resultados


```python
#Datos
E_TNT=4652
Pa=14.7
n=0.5
M=10000
Ec=46.35
P0=5
```

Se calcula la masa equivalente de TNT:

```python
W=n*E_TNT*M/E_TNT
print("W= "+ str(round(W,5))+" Kg")
```

Se calcula la sobrpresión escalada:

```python
Ps=P0/Pa
print("Ps= "+ str(round(Ps,3))+" psi")
```

Se introduce el parámetro z:

```python
z=5.70
print(("z= "+ str(round(z,2))+" m/kg^(1/3)"))
```

Se calcula R:

```python
R=z*W**(1/3)
print(("R= "+ str(round(R,5))+" m"))
```
