# Fracción de flash isoentálpico


## Modelo
Este modelo determina la fracción que se vaporiza de un líquido



## Variables
|Variable|Significado|Unidades|
|------|------|------|
|T|		Temperatura ambiente|K|		
|Tb	|Temperatura normal de ebullición		|K|
|Cp	|	Capacidad calorífical|	KJ/Kg K	|
|hfg|	Calor de vaporización|	KJ/Kg|

### Incógnitas
|Variable|Significado|Unidades|
|------|------|------|
|Fv|		Fracción de masa de líquido vaporizada||		


### Consideraciones
- Los datos de Cp y hfg están en el rango de Temperatura ambiente.





## Desarrollo y ecuaciones

A través de un análisis isoentálpico y asumiendo que el calor sensible contenido en un líquido sobrecalentado a una temperatura por encima de su punto de ebullición es usado para vaporizar  una fracción de el líquido, se obtiene la siguiente ecuación:


- Ecuación 1.  $Fv=Cp\frac{T-Tb}{h_{fg}}$


## Resultados


```python
#Datos
T=298
Tb=231
Cp=2.45
hfg=429
```

Cálculo de la fracción de vaporización.

```python
Fv=Cp*(T-Tb)/hfg
print("Fv = "+ str(round(Fv,3)))
```
