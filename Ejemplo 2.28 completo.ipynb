# Fragmentos expulsados por un Tanque Esférico 


## Modelo
Este modelo estima el número, el  tamaño, la masa y el radio máximo al que salen expulsados los fragmentos de un tanque esférico




## Variables
|Variable|Significado|Unidades|
|------|------|------|

|VR|		Volumen del recipiente			|ft^3|

|P|		Presión de ruptura		|psig|
|$\rho$ R	|	Densidad del material del tanque		|	Lb/ft^3|

|------|------|------|

### Consideraciones 
- El modelo sólo es aplicable para tanques con capacidades  entre 700 y 2500 m^3, es decir el Diámetro debe ser mayor 11m (36 ft) y menor a 17m (56 ft).
- El tanque se fragmenta totalmente.
- Los fragmentos no giran y son de igual tamaño y por lo tanto de igual velocidad y masa.












## Desarrollo y ecuaciones

Se tiene la siguiente relación para calcular el número de fragmentos:

- Ecuación 1. $N_f=-3.77+0.0096V_T $

donde:

-Ecuación 2. $V_T=\frac{\pi D^3_T}{6}[=]m^3$

Para tanques presurizados se tiene:

-Ecuación 3. $u=2.05\sqrt{\frac{P D^3_f}{M_f}}$

Para determinar Mf:

Se calcula el volumen de la pared del tanque (Volumen del recipiente):

- Ecuación 4. $V_R=\frac{\pi}{6}(D_2^3-D_T^3)$

Se calcula la masa del recipiente:


-Ecuación 5. $M_R=V_R\rho_R$

Con estos datos se calcula Mf:

-Ecuación 6. $M_f=\frac{M_R}{N_f}$

Para determinar ${D_f}$
- Se calcula el área del recipiente.
  - Ecuación 7. $A_R= \pi D_T^2$
- Se calcula el área por fragmento
  - Ecuación 8.$A_f= \frac{A_R}{N_f}$
- Se calcula el Df
  - Ecuación 9.$D_f= \sqrt{\frac{A_f}{\pi}}$

De un análisis físico se observa que la máxima distancia teórica es en un ángulo de 45º:

- Ecuación 10. $r_{max}= \frac{u^2}{g}$

## Resultados

```python
g=32.17
DT=50
E=0.0625
P=364.73
rhoR=486.95
```

Capacidad del tanque

```python
import numpy as np
VT=np.pi*DT**3/6
print("VT= "+ str(round(VT,2)) +" ft^3")
```

Número de fragmentos 

```python
Nf=-3.77+0.0096*VT/35.31
print("Nf= "+ str(round(Nf,2)) +" ")
```

Volumen del Recipiente

```python
D2=DT+E
VR=(np.pi/6)*(D2**3-DT**3)
print("VR= "+ str(round(VR,2)) +" ft^3 ")
```

Masa del recipiente

```python
MR=VR*rhoR
print("MR= "+ str(round(MR,2)) +" lb")
```

Masa de cada fragmento

```python
Mf=MR/Nf
print("Mf= "+ str(round(Mf,2)) +" lb")
```

Area del Recipiente

```python
AR=np.pi*DT**2
print("AR= "+ str(round(AR,2)) +" ft^2")
```

Area del fragmento

```python
Af=AR/Nf
print("Af= "+ str(round(Af,2)) +" ft^2")
```

Diámetro del fragmento

```python
Df=np.sqrt(Af/np.pi)
print("Df= "+ str(round(Df,2)) +" ft")
```

Velocidad de los fragmentos

```python
u=2.05*np.sqrt(((P-14.7)*(Df*12)**3)/Mf)
print("u= "+ str(round(u,2)) +" ft/s")
```

Distancia máxima de los fragmentos

```python
rmax=u**2/g
print("r max= "+ str(round(rmax,2)) +" ft")
```
