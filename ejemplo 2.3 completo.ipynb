### Consideraciones
1. No hay diferencia de presión entre ambos puntos, la presión la genera la bomba.
2. La energía debido a la bomba es negativa en el balance de Bernoulli debido a que es energía externa al fenómeno de flujo.

### Desarrollo y ecuaciones
1. Calculamos primeramente el área del orificio (Ecuación 1): $A= \frac{1}{4} \pi d^2$  

2. Se tiene la ecuación de balance de Energia (Ecuación 2) : $\frac{P_2-P_1}{p}+\frac{g}{g_c}(Z_2-Z_1)+\frac{1}{2g_c}(V_2^2-V_1^2)+\sum ef+\frac{W_s}{m}=0$         
    - Se considera:
    - V1=0    (Porque está en reposo)
    - Z2=0    (Nivel de suelo)




3. Se simplifica la ecuación y se multiplica por gc (Ecuación 3) : $-gZ_1+ \frac{V_2^2}{2}+g_c\sum ef-g_c\frac{W_s}{m}=0$


4. Las pérdidas por fricción se cuantifican con la ecuación. (Ecuación 4): $e_f=K_f(\frac{V_2^2}{2g_c})$

5. Para conexiones,válvulas y obstrucciones al flujo se empleea la siguiente ecuación. (Ecuación 5a):$K_f= \frac{K_i}{Re}+K_\infty(1+\frac{24.4}{ID_mm})No_acc $
   - Para entradas y salidas se ha preestablecido que la pérdida de cabeza es 0.5 y 1.0 respectivamente. Pero se puede calcular con la ecuación derivada de la anterior pues teóricamente el diámetro de éstas son infinitos por ello queda: (Ecuación 5b) :$K_f=\frac{K_i}{Re}+K_\infty$

6. Debido a la tubería se tiene la ecuación: (Ecuación 6): $K_f=\frac{4fL}{D}$

7. Para lo anterior se dispone de las siguientes ecuaciones para el Re y el factor de fricción. (Ecuación 7) $Re=\frac{v p D}{\mu}$
   -  Ecuacion 8:$\frac{1}{\sqrt{f}}=-4log_{10}(\frac{\epsilon/D}{3.7065}-\frac{5.0452logA}{Re})$
   -  Ecuacion 9:$A=\frac{\epsilon/D^{1.1098}}{2.8257}+(\frac{7.149}{Re})^{0.8981}$

8. Por último, el flujo másico se determina por (Ecuación 10 ): $m=\rho vA$
