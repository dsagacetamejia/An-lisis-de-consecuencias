# Vaporización


## Modelo
Este modelo calcula la profundidad, velocidad de vaporación y el flux evaporado de un derrame de LNG en un dique.



## Variables
|Variable|Significado|Unidades|
|------|------|------|
|T|		Temperatura del líquido en el charco|		K|
|Tg		|Temperatura del sólido			|K|
|t|		Tiempo de derrame	|		s|
|r		|Radio del dique			|m|
|V|		Volumen derramado|			m^3|
|$\alpha$	|	Difusividad térmica del sólido	|		m^2/s|
|k|		Conductividad térmica del sólido|			W/m*K|
|L		|Calor de vaporización del charco			|J/kg|


### Incógnitas
|Variable|Significado|Unidades|
|------|------|------|
|Ve|		Velocidad de evaporación total|		kg/s|
|qg		|Flux de calor del suelo		|J/s*m^2|
|mf|		Flujo de vaporación|		kg/s*m^2|
|A		|Area		|m^2|


### Consideraciones
- El calor de vaporización del charco es un dato tomado a la temperatura del líquido en el charco (ver Shaw and Briscoe, 1978).
- El flux de calor total dentro del charco es igual al flux de calor del dique de concreto.






## Desarrollo y ecuaciones

Similarmente a una tubería se puede calcular el área de un charco con la ecuación:


- Ecuación 1.  $A=\pi r^2$


La profundidad del charco se puede calcular fácilmente con:

- Ecuación 2. $P_{pool}=\frac{V{Material}}{A}$

La transferencia de calor de la superficie del suelo se modela con la ecuación de flux 

-Ecuación 3.  $q_g=\frac{K_s(T_g-T)}{(\pi a_st)^{0.5}}$

El flux de vaporación se cuantifica con la ecuación:

- Ecuación 4. $m_f=\frac{H}{L}$

Donde H es el flux de calor total dentro del charco que en este caso resulta ser igual a  qg.
Por último la velocidad de evaporización total se calcula con la ecuación:

- Ecuación 5. $V_e=m_f*A$

## Resultados


```python
#Datos
T=109
Tg=293
t=10
r=40
V=1000
alfa=0.000000416
K=0.92
L=498000
```

Cálculo del área del charco

```python
import numpy as np
A=np.pi*r**2
print("Área= "+ str(round(A,3))+"m^2")
```

Profundidad del charco formado

```python
Ppool=V/A
print("Profundidad del charco= "+ str(round(Ppool,3))+" m")
```

Determinación del flux de calor en la superficie del dique

```python
qg=K*(Tg-T)/(np.sqrt(np.pi*alfa*t))
print("Flux de calor en superficie del dique (qg)= "+ str(round(qg,3))+" J/s*m^2")
```

Por tanto, el flux de vaporización

```python
mf=qg/L
print("Flux de vaporación (mf)= "+ str(round(mf,3))+" Kg/s*m^2")
```

Y así la velocidad de evaporación del charco

```python
Ve=mf*A
print("Velocidad vaporación (Ve)= "+ str(round(Ve,3))+" Kg/s")
```
