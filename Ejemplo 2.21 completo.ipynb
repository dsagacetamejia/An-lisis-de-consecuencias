# Modelo de TNO (multienergía). 


## Modelo 
Este modelo determina la sobrepresión y la duración de una explosión de propano confinado en un tanque en función de la distancia.






## Variables
|Variable|Significado|Unidades|
|------|------|------|
|R|		Distancia de reserva		|	m|

|Po|		Presión ambiental		|	Pa|

## Incógnitas


|Variable|Significado|Unidades|
|------|------|------|
|Et	|	Energía total|		J|

|DPs|		Sobrepresión escalada de Sach||		
|Ps	|	Sobrepresión de la explosión|		Pa|
|td0|		Duración escalada de Sach||		
|td|		Tiempo de duración 		|s|




### Consideraciones
-La concentración de vapor en la nube se asume a concentración estequiométrica.

-La fuerza destructiva escogida fue de 7 siendo representativa de éste tipo de explosión.

-El volumen de la nube se asume de 2094 m3 representando el volumen debajo del tanque.

-El calor de combustión de una mezcla estequiométrica de hidrocarburo-aire es 3.5 MJ/m3.









## Desarrollo y ecuaciones

Este modelo asume que el modelamiento  de explosión de una combustión del tipo deflagración es aproximadamente conservativo.

La base para este modelo es que la energía de explosión es altamente dependiente del nivel de congestión y no del nivel de combustible en la nube.

El procedimiento para emplear este método a una explosión de una nube de vapor está dado por los siguientes pasos:



1)  Realizar un análisis de dispersión para determinar la extensión de la nube.

2) Identificar áreas de congestión.

3) Identifique fuentes potenciales de explosión presentes dentro del área que cubre la nube flamable.

4) Estime la energía equivalente decombustible-aire, considerando por separado cada fuente  de explosión, asumiendo que toda mezcla contribuye a la explosión; con el fin de estimar el volumen de combustible-aire y calcular la energía de combustión total.

5)  Estime la fuerza destructiva tomando un estimado que va de 1 a 10, sabiendo que 10 representa una detonación y que 7 es el más representativo.

6)  Una vez determinado lo anterior, a continuación se estiman la sobrepresión y el tiempo de  duración de la explosión en función de la distancia, por medio de unos factores escalados de Sach.

Ver gráficos


Dichas formas escaladas se obtienen con la ayuda de unos gráficos de tipo experimental, para después poder obtener los parámetros buscados en el punto anterior.

Cuyas fórmulas clave son las siguientes:

Para determinar la distancia de la carga (combustible-aire):



- Ecuación 1.  $R=R_{ESCALADO}*(\frac{E}{P_0})^{1/3}$

La sobrepresión de la explosión:


- Ecuación 2.  $P_S=(\Delta P_S)^{ESCALADO}*P_0$

Y por último la duración:

- Ecuación 3.  $P_s=t_d^{ESCALADO}*[\frac{(\frac{E}{P_0})^{1/3}}{v}]$

## Resultados


```python
#Datos

R=30 #m
v= 344 #m/s
P0= 101325 #Pa
V= 2094 #m^3
E= 3500000 #J/m^3
```

Obteniendo energía total considerando la fuente de explosión:

```python
Et=V*E
print("Et= "+ str(round(Et,6))+" J")
```

Obteniendo distancia escalada para utilizar el gráfico correspondiente:

```python
import numpy as np
Rs=R/((Et/P0)**(1/3))
print("Rs= "+ str(round(Rs,2)))
```

Teniendo los datos anteriores y los criterios dados, obtenemos las sobrepresiones: 

```python
DP=0.751
Ps=P0*DP
print("Ps= "+ str(round(Ps,2))+" Pa")
```

De la misma forma, obtenemos ahora las duraciones:

```python
td0=0.268
```

```python
td=(td0*(Et/P0)**(1/3))/v
print("td= "+ str(round(td,4))+" s ")
```
